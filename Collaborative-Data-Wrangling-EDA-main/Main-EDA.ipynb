```python
# Droping missing value seem safer since it‚Äôs just 1 out of many

df_tracks = df_tracks.dropna(subset=['track_name'])
```

```python
# filter the popularity
sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False )
most_popular[:10]
```

```python
df_tracks["duration"]= df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
# keep only numeric columns, then optionally drop some
num_cols = df_tracks.select_dtypes(include='number').columns.difference(['key', 'mode'])
corr_df = df_tracks[num_cols].corr(method='pearson')

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 6))
ax = sns.heatmap(
    corr_df, annot=True, fmt=".1g",
    vmin=-1, vmax=1, center=0, cmap="inferno",
    linewidths=1, linecolor="black"
)
ax.set_title("Correlation Heatmap Between Numeric Variables")
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.tight_layout()
```

Further Steps of Data Cleaning (Performed by: Eva Kaushik)
1. Standardised Column names (lower case + replacing space with underscore)
2. Handling duplicates [Spotify dataset repeat rows (same track_id)]
3. Refine Duration (Instead of seconds, converted to minutes (float, 2 decimals) for easier interpretation)
4. Categorical Consistency: Some columns (key, mode, time_signature) might have unexpected values ‚Äî check their unique values.
5. Saved Cleaned Dataset

```python
#Standardize the column names 
df_tracks.columns = [c.strip().lower().replace(" ", "_") for c in df_tracks.columns]
```

```python
# Handles Duplicate
df_tracks = df_tracks.drop_duplicates(subset="track_id")
```

```python
# Refine duration 
df_tracks["duration_min"] = df_tracks["duration"] / 60
```

```python
#Checking Categorical Consistency 
print(df_tracks["key"].unique())
print(df_tracks["mode"].unique())
print(df_tracks["time_signature"].unique())
```

Exploratory Data Analysis (Performed by: Eva Kaushik)
1. Loading the cleaned data (checked column names)
2. Computing Mean, Median and Correlation
3. Histogram of Track Duration (Frequency vs. Duration)
4. Scatter Plot (Danceability vs. Energy)

```python
import pandas as pd

df = pd.read_csv("/Users/ekaushik/Desktop/Collaborative-Data-Wrangling-EDA/Collaborative-Data-Wrangling-EDA-main/data/spotify_cleaned.csv")
df.head()
```

```python
print(df.columns.tolist())
```

```python
# Computing mean, median, and correlation
mean_duration = df['duration_min'].mean()
median_duration = df['duration_min'].median()
correlation = df[['danceability', 'energy']].corr().iloc[0,1]

print(f"Mean Duration (min): {mean_duration:.2f}")
print(f"Median Duration (min): {median_duration:.2f}")
print(f"Correlation (Danceability vs Energy): {correlation:.2f}")
```

```python
#Histogram of Track Duration 
plt.figure(figsize=(10,6))
sns.histplot(df['duration_min'], bins=30, kde=True)
plt.title("Distribution of Track Duration (minutes)")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")
plt.show()
```

```python
#Scatterplot 
plt.figure(figsize=(10,6))
sns.scatterplot(x='danceability', y='energy', data=df)
plt.title("Danceability vs Energy")
plt.xlabel("Danceability")
plt.ylabel("Energy")
plt.show()
```

```python
# Correlation Heatmap
numeric_df = df.select_dtypes(include='number')
plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
